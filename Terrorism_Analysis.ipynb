```python
df_iraq.reset_index(drop=True, inplace= True)
df_iraq['Killed'].fillna(0, inplace = True )
df_iraq['Wounded'].fillna(0, inplace = True )
```

```python
kill= []
wound = []
year = []
s=0
c=0
for i in range(0, len(df_iraq)-1, 1):

    if(df_iraq['Year'][i]!=df_iraq['Year'][i+1]):
        s = s+ df_iraq['Killed'][i]
        c = c + df_iraq['Wounded'][i]
        kill.append(s)
        wound.append(c)
        year.append(df_iraq['Year'][i])
        s=0
        c=0
    else:
        s=s+df_iraq['Killed'][i]
        c = c + df_iraq['Wounded'][i]
```

```python
kill = np.array(kill)
year = np.array(year)
```

```python
plt.plot(year, kill, marker='.', label= "Killed")
plt.plot(year, wound, marker='.', label= "Wounded")
plt.xlabel("Year")
plt.ylabel("Kills")
plt.title("Total Casualities in Iraq")
plt.legend()
plt.grid()
plt.show()


plt.bar(year, wound, label= "Wounded")
plt.bar(year, kill, label= "Killed")
plt.xlabel("Year")
plt.ylabel("Kills")
plt.title("Total Casualities in Iraq")
plt.legend()
plt.grid()
plt.show()
```
